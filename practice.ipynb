```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(
    "https://en.wikipedia.org/wiki/Tata_Consultancy_Services",
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(id="mw-content-text")
    )
)

webdata = loader.load()

print("documents loaded:", len(webdata))

if webdata:
    text = webdata[0].page_content
    print(text[:2000])  # print first 2000 chars
else:
    print("No content was loaded")

sentences = [s.strip() for s in text.split(". ") if s.strip()]
for s in sentences[:10]:
    print(f"- {s}")
```

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(
    "https://en.wikipedia.org/wiki/Tata_Consultancy_Services",
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(id="mw-content-text")
    )
)

webdata = loader.load()

print("documents loaded:", len(webdata))

if webdata:
    text = webdata[0].page_content
    print("✅ Content extracted\n")

    # Split into sentences and print as bullet points
    sentences = [s.strip() for s in text.split(". ") if s.strip()]
    for s in sentences[:10]:   # limit to first 10 sentences
        print(f"- {s}")
else:
    print("No content was loaded")
```
