```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("esp32_datasheet.pdf")
pdfData = loader.load()

print(pdfData[0])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

textSplitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=20
)

finalDocuments = textSplitter.split_documents(pdfData)

print(finalDocuments)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_community.vectorstores import FAISS

db = FAISS.from_documents(finalDocuments, embeddings)

print(db)
```

```python
query = input("Enter your question about the PDF: ")

docs = db.similarity_search(query)

for i, doc in enumerate(docs):
    print(f"\nResult {i+1}")
    print("-" * 50)
    print(doc.page_content)
```
